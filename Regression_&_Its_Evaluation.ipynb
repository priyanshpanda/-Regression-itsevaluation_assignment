```python
''' Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points '''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a Linear Regression model object
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Get the predictions to plot the regression line
Y_pred = model.predict(X)

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Actual Data Points')

# Plot the regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')

# Add titles and labels for clarity
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Print the model coefficients
print(f"Intercept (b0): {model.intercept_}")
print(f"Slope (b1): {model.coef_[0]}")
```

```python
''' Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]'''
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Define the independent variables (X) and the dependent variable (Y)
X = df[['Area', 'Rooms']]
Y = df['Price']

# Create and fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Print the model's intercept and coefficients
print(f"Intercept (b0): {model.intercept_}")
print(f"Coefficient for Area (b1): {model.coef_[0]}")
print(f"Coefficient for Rooms (b2): {model.coef_[1]}")
```

```python
'''Question 8: Implement polynomial regression on the following data: ● X = [1, 2, 3, 4, 5] ● Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# 1. Create Polynomial Features (degree=2)
# This will transform our input X into three columns: X^0, X^1, and X^2
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# 2. Fit the Linear Regression model to the polynomial features
model = LinearRegression()
model.fit(X_poly, Y)

# 3. Generate a smooth curve for plotting
# We create a new set of X values with more points for a smoother line
X_plot = np.linspace(1, 5, 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)
Y_plot = model.predict(X_plot_poly)

# 4. Plot the results
plt.scatter(X, Y, color='blue', label='Actual Data Points')
plt.plot(X_plot, Y_plot, color='red', linewidth=2, label='Polynomial Regression Curve')

# Add titles and labels for clarity
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data: ● X = [10, 20, 30, 40, 50] ● Y = [15, 35, 40, 50, 65] Assess heteroscedasticity by examining the spread of residuals
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# 1. Fit the regression model
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# 2. Calculate the residuals
residuals = Y - Y_pred

# 3. Create the residuals plot
plt.scatter(X, residuals, color='green', edgecolor='black')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2) # Add a zero line
plt.title('Residuals Plot')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```
